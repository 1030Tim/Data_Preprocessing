```python
import pandas as pd
train = pd.read_csv("/Users/juhn/Desktop/stady_file/資料整理/SCoML/train.csv",encoding = "utf-8")
test = pd.read_csv("/Users/juhn/Desktop/stady_file/資料整理/SCoML/test.csv",encoding = "utf-8")
```

```python
test.isna().sum().sum()
```

```python
def Clean_Number_None(df):
    median = df.median(numeric_only=True)
    df = df.fillna(median)
    return df

train = Clean_Number_None(train)
test = Clean_Number_None(test)
```

```python
train.isna().sum()
```

```python
for col in train.columns:
    if (train[col].isna().sum() > 0):
        most = train[col].value_counts().idxmax()
        train[col] = train[col].fillna(most)
train.isna().sum().sum()
```

```python
for col in test.columns:
    if test[col].isna().sum() > 0:
        most = test[col].value_counts().idxmax()
        test[col] = test[col].fillna(most)
```

```python
test.isna().sum().sum()
```

```python
train_Encoder = pd.get_dummies(train["gender"]).astype("int64")
train = pd.concat([train,train_Encoder],axis=1)
train = train.drop("gender",axis=1)

test_Encoder = pd.get_dummies(test["gender"]).astype("int64")
test = pd.concat([test,test_Encoder],axis=1)
test = test.drop("gender",axis=1)
```

```python
Encoder = pd.get_dummies(train["apache_3j_bodysystem"]).astype("int64")
train = pd.concat([train,Encoder],axis=1)
train = train.drop("apache_3j_bodysystem",axis=1)

Encoder = pd.get_dummies(test["apache_3j_bodysystem"]).astype("int64")
test = pd.concat([test,Encoder],axis=1)
test = test.drop("apache_3j_bodysystem",axis=1)

test
```

```python
Encoder = pd.get_dummies(train["icu_type"]).astype("int64")
train = pd.concat([train,Encoder],axis=1)
train = train.drop("icu_type",axis=1)

Enocder = pd.get_dummies(test["icu_type"]).astype("int64")
test = pd.concat([test,Enocder],axis=1)
test = test.drop("icu_type",axis=1)
```

```python
def drop_data(df)->any:
    df = df.drop(["encounter_id","patient_id","hospital_id"],axis=1)
    return df
train = drop_data(train)
test = drop_data(test)
train
```

```python
def OneHotEncoder(df,s):
    Encoder = pd.get_dummies(df[s]).astype("int64")
    df = pd.concat([df,Encoder],axis=1)
    df = df.drop(s,axis=1)
    return df

train = OneHotEncoder(train,"ethnicity")
test = OneHotEncoder(test,"ethnicity")
```
